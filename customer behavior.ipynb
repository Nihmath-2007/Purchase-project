```python
# CODE FOR IMPORT CSV

import pandas as pd
from sqlalchemy import create_engine


# MySQL Connection
engine = create_engine(
    "mysql+pymysql://root:2007@localhost:3306/pda"
)

# Upload CSV to MySQL
df.to_sql(
    name="purchase",
    con=engine,
    if_exists="replace",
    index=False
)

# Check Data
result = pd.read_sql(
    "SELECT * FROM purchase LIMIT 5",
    engine
)

print(result)
```
